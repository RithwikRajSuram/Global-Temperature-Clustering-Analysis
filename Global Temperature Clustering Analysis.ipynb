```python
import pandas as pd
data=pd.read_csv("/content/Data-Week6 (1).csv")
data.head(2)
```

```python
x=data[["Year", "Temp"]]
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
SSE=[]
for k in range(1,10):
  km=KMeans(n_clusters=k)
  km.fit(x)
  SSE.append(km.inertia_)
plt.figure(figsize=(20,10))
plt.plot(range(1,10),SSE,color='b',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
X1=data[['Year']].values
X2=data[['Temp']].values
km=KMeans(n_clusters=3)
km.fit(X2)
pred=km.predict(X2)
plt.scatter(X1,X2,c=pred,s=50,cmap='viridis')
centers=km.cluster_centers_
```
